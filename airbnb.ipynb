```python
# Handling null values in Host dataframe
df_host["host.host_location"].fillna("Not Specified",inplace=True)

df_host["host.host_response_time"].fillna("Not Specified",inplace=True)
df_host["host.host_response_rate"].fillna("Not Specified",inplace = True)
df_host["host.host_neighbourhood"]= df_host["host.host_neighbourhood"].replace({'':"Not Specified"})
df_host["host.host_neighbourhood"].fillna("Not Specified",inplace = True)
```

```python
df_host.isnull().sum()
```

```python
# Changing the "True" or "False" features

df_host["host.host_is_superhost"]= df_host["host.host_is_superhost"].map({False: "No", True: "Yes"})
df_host["host.host_has_profile_pic"]= df_host["host.host_has_profile_pic"].map({False: "No", True: "Yes"})
df_host["host.host_identity_verified"]= df_host["host.host_identity_verified"].map({False: "No", True: "Yes"})
```

```python
df_address = df.iloc[:,137:147]
df_address = df_address.rename(columns={'address.location.coordinates[0]':'longitude','address.location.coordinates[1]':'latitude'})
df_address.head(2)
```

```python
df_address.isnull().sum()
```

```python
#handling empty values and replace them with not specified
df_address1 = df_address.iloc[:,:-1]
df_address1.replace('', np.nan, inplace=True)
df_address1.fillna('Not Specified', inplace=True)

df_address1["address.location.is_location_exact"]=df_address["address.location.is_location_exact"]


df_address1.head(2)
df_address1["address.location.is_location_exact"]
```

```python
df_address1["address.location.is_location_exact"]= df_address1["address.location.is_location_exact"].map({False:"No", True:"Yes"})
df_address1.insert(0,'_id',df['_id'])
df_address1
```

```python
df_availability = df[['_id','availability.availability_30','availability.availability_60','availability.availability_90','availability.availability_365']]
df_availability
```

```python
df_availability.isnull().sum()
```

```python
# Merging All data frames

df_final = pd.merge(df1, df_host, on='_id')
df_final = pd.merge(df_final,df_address1, on='_id')
df_final= pd.merge(df_final,df_availability,on='_id')

df_final.head(2)
```

```python
# Convert final dataframe as csv file.

df_final.to_csv('Airbnb_analysis_finaldata', index = False)
```
